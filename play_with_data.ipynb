```python
def get_post_creator(post):
    for entry in post['change_log']:
        if entry['type'] == 'create':
            return entry['uid']
        
def get_posts_by_student(student_id):
    student_posts = []
    with open('csc108_fall2021_json.txt', 'r') as f:
        all_posts = json.load(f)
        for p in all_posts:
            if get_post_creator(p) == student_id:
                student_posts.append(p)
    return student_posts


#csc108.get_post('kxxg3uqr8kc4ui')
#get_posts_by_student('ksoq6sru6w34ak')
```

### Get statistics as outlined at the start of the notebook

```python
from tqdm import trange, tqdm
#feed = csc108.get_feed()

with open('csc108_fall2021_json.txt', 'r') as f:
    posts = json.load(f)

#posts = csc108.iter_all_posts()

# helpful / total_q's

students_helpful = {} # dict of students that have marked a response as helpful for a question they asked 
students_total = {} # total number of questions asked by the student

num_posts = len(posts) # total number of posts
num_questions = 0 # total number of questions
num_helpful = 0 # total number of questions that have recvd a helpful response (not necessarily from the question asker)


for post in tqdm(posts):    
    if post['type'] == 'question':
        num_questions += 1
        
        post_creator = get_post_creator(post)
        if post_creator not in students_total:
            students_total[post_creator] = 1
            students_helpful[post_creator] = 0
        else:
            students_total[post_creator] += 1
            
        marked_helpful = False
        asker_marked_helpful = False
        
        for c in post['children']:
            if asker_marked_helpful:
                break
            if c['type'] == 's_answer' or c['type'] == 'i_answer': # treating as 1 for now
                # someone marked the question as helpful (not necessarily the question asker)
                if len(c['tag_endorse_arr']) > 0 and not marked_helpful:
                    marked_helpful = True
                    num_helpful += 1
                    
                # all users that marked the response as helpful
                for s in c['tag_endorse_arr']:
                    # only add student if they are the question asker
                    if s == post_creator:
                        students_helpful[s] += 1
                        #print('adding student {}'.format(s))
                        asker_marked_helpful = True
                        break # want to break out of children as well


                        
                
            
#         children = post['children']
#         text = json.dumps(post['children'][1], sort_keys=True, indent=4)
#         print(text)
#         #print(post.keys())
#         #print(post)
#         #print(post['history'])
#         #print(post['tag_good'])
#         #print(len(post['tag_good']))
       
#         break
  
    
```
